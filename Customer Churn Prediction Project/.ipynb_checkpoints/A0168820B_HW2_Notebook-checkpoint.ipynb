## Data Manipulation

```python
## convert all missing data to nan and drop na
df.replace(" ", np.nan, inplace=True)
df.dropna(inplace=True)

#convert to numeric (float) type
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"])
```

```python
## discretize Tenure
# define the bins
bins = [0, 12, 24, 36, 48, 60, 72]

# cut the series into bins
tenure_binned = pd.cut(df["tenure"], bins)
print(tenure_binned.value_counts())

df["tenure_bins"] = tenure_binned.apply(lambda x: str(x))
```

#### Observations:
From the above visualisations, we can observe that:
1. Tenure: Churn is high when the tenure is low. This is probably due to the fact that the customers may still be deciding on their preferrred telco. It is only after the 18-months barrier that the churn rate will decrease and customers will stay loyal to one telco.

2. MonthlyCharges: Churn is high when monthly charges is high and churn is low when monthly charges are low. This implies that MonthlyCharges is positively correlated to Churn. This could be because customers who are willing to pay tend to demand better service and are easily dissatisfied, prompting shifts to other services

3. TotalCharges: There are too many people with TotalCharges lower than 2000. There is hence not enough information to infer the correlation between TotalCharges and Churn Rate.

## Data Processing

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
## drop tenure bins as we do not need it anymore
df.drop(columns=["tenure_bins"], inplace=True)
cat_features.remove("tenure_bins")
```

```python
## get list of all types of column names

#customer id col
Id_col     = ['customerID']

#Target columns
target_col = ["Churn"]

#categorical columns
cat_cols   = cat_features

#numerical columns
num_cols   = cts_features

#Binary columns with 2 values
bin_cols   = df.nunique()[df.nunique() == 2].keys().tolist()

#Columns more than 2 values
multi_cols = [i for i in cat_cols if i not in bin_cols]

#Label encoding Binary columns ==> make it into 1s and 0s
le = LabelEncoder()
for i in bin_cols :
    df[i] = le.fit_transform(df[i])
    

#Duplicating columns for multi value columns
df = pd.get_dummies(data = df,columns = multi_cols)
```

```python
#Scaling Numerical columns
std = StandardScaler()
scaled = std.fit_transform(df[num_cols])
scaled = pd.DataFrame(scaled,columns=num_cols)

#dropping original values merging scaled values for numerical columns
df_original = deepcopy(df)
df.drop(columns = num_cols,axis = 1, inplace=True)

# Reset the index
df.reset_index(drop=True, inplace=True)
scaled.reset_index(drop=True, inplace=True)
df = df.merge(scaled,left_index=True,right_index=True,how = "left")
```

```python
## calculate scores to check for significance of all features
scores = chi2(df_original.iloc[:, 1:-1], churn)[0]

# iterate over scores and see which ones are highest
scores_list = []
for col_num, score in enumerate((scores)):
    colname = df_original.columns[col_num+1]
    scores_list.append((colname, score))

scores_list.sort(key=lambda tup: tup[1])
scores_list
```

#### Prepare the dataframe

```python
## shift churn column to the right
churn = df["Churn"]
df.drop("Churn", inplace=True, axis=1)
df = df.join(churn)
df.head()
```

```python
## create a copy of df specifically for decision tree
df_decision_tree = deepcopy(df)

## select top 6 features
top_features  = score[score["feature_type"] == "Categorical"]["features"][:3].tolist()
top_features += ["MonthlyCharges", "tenure"]
print(top_features)

## make our dataset by selecting the top features
df_decision_tree = df_decision_tree[top_features + ["Churn"]]
df_decision_tree.head()
```

#### Train - Test Split using oversampling

```python
# Split train data: 70% for model fit, 30% for validation
X = df_decision_tree[top_features]
y = df_decision_tree["Churn"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12345)
```

```python
training_set = X_train.merge(pd.DataFrame(y_train), left_index=True,right_index=True,how = "left")
```

```python
# Address the Unbalanced dataset by Oversampling
random.seed(12345)

# Class count
count_class_0, count_class_1 = training_set["Churn"].value_counts()

# Divide by class
df_class_0 = training_set[training_set["Churn"] == 0]
df_class_1 = training_set[training_set["Churn"] == 1]

# Address unblanced data issue: Oversampling on data samples with rare outputs (which is '1' in this example)
df_class_1_over = df_class_1.sample(count_class_0, replace=True, random_state=12345)
training_set = pd.concat([df_class_1_over, df_class_0], axis=0)

training_set["Churn"].value_counts()
```

```python
## split back the X_train and y_train
X_train, y_train = training_set.iloc[:, :-1], training_set["Churn"]
```

#### Train the decision tree model

```python
# Fit the model on train data
decision_tree = DecisionTreeClassifier(criterion="gini", random_state=12345)
decision_tree_model = decision_tree.fit(X_train, y_train)
decision_tree_model.classes_
```

```python
# Get predicted labels for test data
y_pred = decision_tree_model.predict(X_test)
```

#### Evaluate Decision Tree Performance

```python
model_metrics = {}

# Performance of decision tree model
print ("Accuracy: ", accuracy_score(y_test, y_pred))
print ("Sensitivity: ", recall_score(y_test, y_pred))
print ("Precision: ", precision_score(y_test, y_pred))
print("f1 Score:", f1_score(y_test, y_pred))
print("Recall Score", recall_score(y_test, y_pred))

decision_tree_metrics = pd.DataFrame({"Model": ["Decision Tree"],
                                      "Accuracy_score"  : [accuracy_score(y_test, y_pred)],
                                      "Recall_score"    : [recall_score(y_test, y_pred)],
                                      "Precision"       : [precision_score(y_test, y_pred)],
                                      "f1_score"        : [f1_score(y_test, y_pred)],
                                      "Area_under_curve": [roc_auc_score(y_test,y_pred)]
                                     })

model_metrics["Decision Tree"] = decision_tree_metrics
```

```python
## get the feature importance
coefficients  = pd.DataFrame(decision_tree.feature_importances_)
column_df     = pd.DataFrame(top_features)
coef_sumry    = (pd.merge(coefficients,column_df,left_index= True,
                          right_index= True, how = "left"))
coef_sumry.columns = ["coefficients","features"]
coef_sumry    = coef_sumry.sort_values(by = "coefficients",ascending = False)


#confusion matrix
conf_matrix = confusion_matrix(y_test,y_pred)
#roc_auc_score
model_roc_auc = roc_auc_score(y_test,y_pred) 
print ("Area under curve : ",model_roc_auc,"\n")

# Get predicted scores Pr(y=1): Used as thresholds for calculating TP Rate and FP Rate
score = decision_tree.predict_proba(X_test)[:, 1]


fpr,tpr,thresholds = roc_curve(y_test,score)

#plot confusion matrix
trace1 = go.Heatmap(z = conf_matrix ,x = ["Not churn","Churn"],y = ["Not churn","Churn"],
                    showscale  = False,colorscale = "Picnic",name = "matrix")

#plot roc curve
trace2 = go.Scatter(x = fpr,y = tpr,
                    name = "Roc : " + str(model_roc_auc),
                    line = dict(color = ('rgb(22, 96, 167)'),width = 2))
trace3 = go.Scatter(x = [0,1],y=[0,1],
                    line = dict(color = ('rgb(205, 12, 24)'),width = 2,
                    dash = 'dot'))

#plot coeffs
trace4 = go.Bar(x = coef_sumry["features"],y = coef_sumry["coefficients"],
                name = "coefficients",
                marker = dict(color = coef_sumry["coefficients"],
                              colorscale = "Picnic",
                              line = dict(width = .6,color = "black")))

#subplots
fig = tls.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                        subplot_titles=('Confusion Matrix',
                                        'Receiver operating characteristic', 'Feature Importances'))

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
fig.append_trace(trace3,1,2)
fig.append_trace(trace4,2,1)

fig['layout'].update(showlegend=False, title="Model performance" ,
                     autosize = False,height = 900,width = 800,
                     plot_bgcolor = 'rgba(240,240,240, 0.95)',
                     paper_bgcolor = 'rgba(240,240,240, 0.95)',
                     margin = dict(b = 195))
fig["layout"]["xaxis2"].update(dict(title = "false positive rate"))
fig["layout"]["yaxis2"].update(dict(title = "true positive rate"))
fig["layout"]["xaxis3"].update(dict(showgrid = True,tickfont = dict(size = 10),
                                    tickangle = 90))
py.iplot(fig)
```

From the above results, we can see that the `logistic regression` and `Support Vector Machine` models have the highest Accuracy and Recall scores. I would like to explore these models further by performing hyperparameter optimization to enhance their performance. I will be using the SMOTE oversampling technique rather than the earlier oversampling method, just to push the model even further in terms of performance.

### Note to Prof/ TA: Please do not run the below code, as it takes very very very long. I have printed out everything and left it there for your convenience :)

### Logistic Regression

```python
# Initialise the model with default parameters
logistic_regression = LogisticRegression(fit_intercept=True, max_iter=1000, tol=2e-9, random_state=12345)
```

#### Prepare the dataset

```python
df_lr = deepcopy(df)
# Drop gender and phoneservice as their p-value is greater than 0.05 at the 5% significance level
df_lr.drop(["gender", "PhoneService", "TotalCharges"], axis=1, inplace=True)
```

#### Train-Test Split

```python
# Split dataset to do validation: 70% for model fit, 30% for test
X = df_lr[[x for x in df_lr.columns if x not in ["customerID", "Churn"]]]
y = df_lr["Churn"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12345, stratify=y)
```

#### SMOTE Oversampling

```python
# Address the Unbalanced Training Set by Oversampling SMOTE

sm = SMOTE(random_state=12345)
sm_X_train, sm_y_train = sm.fit_sample(X_train, y_train.ravel())
```

#### Determine Hyperparameters

```python
# Create regularization penalty space
penalty = ['l1', 'l2']

# Create regularization hyperparameter space
C = np.logspace(0, 4, 10)

# Create hyperparameter options
hyperparameters = dict(C=C, penalty=penalty)
```

#### Grid Search using k-fold cross validation

```python
grid_search = GridSearchCV(logistic_regression, hyperparameters, cv=5, verbose=0, n_jobs=-1)
best_model = grid_search.fit(sm_X_train, sm_y_train)
```

```python
print(best_model.best_score_)
print(best_model.best_params_)
print(best_model.best_estimator_)
print('Best Penalty:', best_model.best_estimator_.get_params()['penalty'])
print('Best C:', best_model.best_estimator_.get_params()['C'])
```

#### Redefine the model with the new hyperparameters

```python
lr_final = LogisticRegression(C=1, class_weight=None, dual=False,
          fit_intercept=True, intercept_scaling=1, max_iter=1000,
          multi_class='ovr', n_jobs=1, penalty='l2', random_state=12345,
          solver='liblinear', tol=2e-09, verbose=0, warm_start=False)

lr_final.fit(sm_X_train, sm_y_train)
y_pred = lr_final.predict(X_test)
```

#### Model Performance

```python
print ("Accuracy: ", accuracy_score(y_test, y_pred))
print ("Sensitivity: ", recall_score(y_test, y_pred))
print ("Precision: ", precision_score(y_test, y_pred))
print ("f1 Score: ", f1_score(y_test, y_pred))
print("ROC_AUC_Score:", roc_auc_score(y_test, y_pred))
```
